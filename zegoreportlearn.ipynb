```python
# Importing useful libraries

import pandas as pd
import numpy as np
import seaborn as sns
import sys
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import plotly.express as px
```

```python
# Loading data to dataset

dataset = pd.read_excel(
    "./First Week of March.xlsx")
dataset.columns = dataset.iloc[0]
dataset = dataset[1:]
dataset
```

```python
dataset["Initiation Date"] = pd.to_datetime(
    dataset["Initiation Date"], format="ISO8601")
dataset
```

```python
# Adding month and year and set up previsors and aimed results

dataset['month'] = dataset['Initiation Date'].dt.month
dataset['year'] = dataset['Initiation Date'].dt.year

previsors = dataset.iloc[:, [ 12, 13, 16, 17]].values
results = dataset.iloc[:, 15].values

print(previsors, results)
```

```python
# Convertendo o que for string para alguma coisa relacionada a um numero

labelencoder_teste = LabelEncoder()
previsors[:, 1] = labelencoder_teste.fit_transform(previsors[:, 1])

previsors
```

```python
# Criando um escalonamento dos dados para que possa ser feito os calculos nos nossos algoritmos

scaler_data = StandardScaler()
previsors = scaler_data.fit_transform(previsors)
previsors[0]
```

```python
# Divinding the bases
from sklearn.model_selection import train_test_split

previsors_test, previsors_train, classe_test, classe_train = train_test_split(
    previsors, results, test_size=0.60, random_state=0)
previsors_train.shape, classe_train.shape
```

```python
# Saving the database
import pickle

with open('./zegodata.pkl', mode='wb') as f:
    pickle.dump([previsors_train, classe_train], f)

with open('./zegotestdata.pkl', mode='wb') as arquivo:
    pickle.dump([previsors_test, classe_test], arquivo)
```

```python
from sklearn.naive_bayes import GaussianNB

naive_recurrence = GaussianNB()
naive_recurrence.fit(previsors_train, classe_train)
prevision = naive_recurrence.predict(previsors_test)
sns.countplot(x=prevision)
```

```python
# Verifying accuracy

from sklearn.metrics import accuracy_score, confusion_matrix
from yellowbrick.classifier import ConfusionMatrix
confusion_matrix(classe_test, prevision)
cm  = ConfusionMatrix(naive_recurrence)
cm.fit(previsors_train, classe_train)
cm.score(previsors_test, classe_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(classe_test, prevision))
```

```python
# Arvores de decisão

from sklearn.tree import DecisionTreeClassifier

# quando utilizamos o objeto DecisionTreeClassifier, temos que definir o criterion de entropia e realizar novamente o aprendizado com o .fit

decisiontree = DecisionTreeClassifier(criterion='entropy')
decisiontree.fit(previsors_train, classe_train)
```

```python
previsao = decisiontree.predict(previsors_train)
previsao

from sklearn.metrics import accuracy_score, confusion_matrix
from yellowbrick.classifier import ConfusionMatrix
confusion_matrix(classe_test, previsao)
cm  = ConfusionMatrix(naive_recurrence)
cm.fit(previsors_train, classe_train)
cm.score(previsors_test, classe_test)
```
